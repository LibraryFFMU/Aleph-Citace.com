# Příprava dat Citace.com

```python
Citace_1 = pd.read_excel("citace.xlsx")
Citace_2 = pd.read_excel("citace(1).xlsx")
```

```python
#spojení datasetů do jednoho
Citace_com_all = pd.concat([Citace_1, Citace_2])
```

```python
#seřazeno dle počtu
Citace_com_all.groupby(["Druh"]).size().sort_values(ascending=False)
```

```python
#příprava df s vybranými sloupci
Citace_com = Citace_com_all[["Druh", "Název", "Datum vydání", 
                            "Nakladatelství", "ISBN", "ISSN", "DOI", "Kód dokumentu"]]

Citace_vyber = Citace_com[Citace_com["ISBN"].notnull() | Citace_com["ISSN"].notnull()
                        | Citace_com["DOI"].notnull()]
Citace_vyber.head()
Citace_vyber.to_excel("Citace_com_vyber.xlsx")
```

```python
#příprava ISBN a ISSN

# ISBN - mažeme "-" a "978", mažeme poslední kontrolní znak
Citace_vyber['ISBN_norma'] = Citace_vyber['ISBN'].str.replace('-','').str.replace('–','').str.replace(r'^978', '')
Citace_vyber['ISBN_norma'] = Citace_vyber.ISBN_norma.str[:-1]

#ISSN - mažeme "-"
Citace_vyber['ISSN_norma'] = Citace_vyber['ISSN'].str.replace('-','').str.replace('–','')

# #ISBN_clear.to_csv('ISBN_2022_FF.csv', encoding='utf-8', index=False)
Citace_vyber.head()
```

# Příprava dat za knihovnu
- data z discovery
     - exktrakce ISBN a ISSN se souboru hodings - export z EDS
- data z knihovního systému Aleph
    - připravené soubory ISBN_aleph.csv | ISSN_aleph.csv

```python
#ISSN, ISBN z discovery systému

holdings = pd.read_csv('complete-holdings-MU-20220401.csv', sep= ',')
holdings.info()
holdings.head()
```

## ISBN 

```python
PrintISBN = holdings[["KBID", "Title","PrintISBN"]]
PrintISBN.rename(columns={"PrintISBN":"ISBN", "Title":"nazev"}, inplace=True)

OnlineISBN = holdings[["KBID","Title","OnlineISBN",]]
OnlineISBN.rename(columns={"OnlineISBN":"ISBN", "Title":"nazev"}, inplace=True)

discovery_ISBN = pd.concat([PrintISBN, OnlineISBN], axis=0)
discovery_ISBN.dropna(inplace=True)

discovery_ISBN.info(show_counts = True)
```

```python
discovery_ISBN["ISBN_clear"] = discovery_ISBN["ISBN"].str.extract("^([0-9xX-]+)")
discovery_ISBN["ISBN_clear"] = discovery_ISBN["ISBN_clear"].str.replace("-","").str.replace(r'^978', '')
discovery_ISBN = discovery_ISBN[discovery_ISBN.ISBN_clear.apply(lambda x: len(str(x))== 10 or (len(str(x))== 13 and x.startswith('979')))]
discovery_ISBN["ISBN_clear"] = discovery_ISBN.ISBN_clear.str[:-1]

discovery_ISBN.rename(columns = {"KBID" : "ID", "ISBN_clear" : "norma"}, inplace=True)

discovery_ISBN.drop_duplicates(subset=["norma","nazev"], keep="first", inplace=True)

discovery_ISBN["zdroj"]= "discovery"
discovery_ISBN.head()
```

```python
#změna pořadí sloupců - příprava na spojení dat z Alephu
discovery_ISBN = discovery_ISBN.reindex(columns=["ID", "ISBN", "nazev", "norma", "zdroj"])
discovery_ISBN.info()
```

```python
#MU ISBN
aleph = pd.read_csv('Isbn_aleph.csv', dtype=str)
aleph.rename(columns = {'sysno':'ID'}, inplace = True)
aleph["zdroj"] = "aleph"
aleph.head()
```

```python
ISBN_MU = pd.concat([discovery_ISBN,aleph])
ISBN_MU.to_csv("ISBN_MU.csv", index = False)
ISBN_MU.info()
```

```python
#počet ISBN z alephu a discovery
ISBN_MU.groupby(["zdroj"]).size()
```

## ISSN

```python
PrintISSN = holdings[["KBID", "Title","PrintISSN"]]
PrintISSN.rename(columns={"PrintISSN":"ISSN"}, inplace=True)

OnlineISSN = holdings[["KBID","Title","OnlineISSN",]]
OnlineISSN.rename(columns={"OnlineISSN":"ISSN"}, inplace=True)

discovery_ISSN = pd.concat([PrintISSN, OnlineISSN], axis=0)
discovery_ISSN.dropna(inplace=True)

#PrintISSN.info()
#OnlineISSN.info()
discovery_ISSN.info(show_counts = True)
```

```python
discovery_ISSN["ISSN_clear"] = discovery_ISSN["ISSN"].str.extract("^([0-9xX-]+)")

discovery_ISSN["ISSN_clear"] = discovery_ISSN['ISSN_clear'].str.replace('-','')
discovery_ISSN = discovery_ISSN[discovery_ISSN.ISSN_clear.apply(lambda x: len(str(x))== 8 )]

discovery_ISSN.rename(columns = {"KBID" : "ID", "Title": "nazev", "ISSN_clear": "norma"}, inplace=True)

discovery_ISSN.drop_duplicates(subset=["norma","nazev"], keep="first", inplace=True)

discovery_ISSN["zdroj"]= "discovery"
discovery_ISSN.info()
```

```python
#změna pořadí sloupců - příprava na spojení dat z Alephu
discovery_ISSN = discovery_ISSN.reindex(columns=["ID", "ISSN", "nazev", "norma", "zdroj"])
discovery_ISSN.info()
```

```python
aleph = pd.read_csv('ISSN_aleph.csv', dtype=str)
aleph.rename(columns = {'sysno':'ID'}, inplace = True)
aleph["zdroj"] = "aleph"
aleph.head()
```

```python
ISSN_MU = pd.concat([discovery_ISBN,aleph])
ISSN_MU.to_csv("ISSN_MU.csv", index = False)
ISSN_MU.info()
```

```python
#počet ISSN aleph, discovery
ISSN_MU.groupby(["zdroj"]).size()
```

# Srování Citace.com, fond Masarykovi univerzity

```python
ISBN_result = pd.merge(Citace_vyber, ISBN_MU[["norma","zdroj"]], left_on = "ISBN_norma", right_on = "norma", how="left")
ISBN_result.head()
```

```python
#kolik je nalezených shod
ISBN_result.groupby(["zdroj"]).size()
```

```python
#co máme ve fondu
ISBN_notNan = ISBN_result[ISBN_result["zdroj"].notnull() & ISBN_result["ISBN"].notnull()]
ISBN_mame = ISBN_notNan.groupby(["norma", "Název"]).size().sort_values(ascending=False).to_frame().reset_index()

ISBN_mame.info()
ISBN_mame.head(20)
```

```python
#co nemáme ve fondu
ISBN_Nan = ISBN_result[ISBN_result["zdroj"].isna() & ISBN_result["ISBN"].notnull()]
ISBN_nemame = ISBN_Nan.groupby(["ISBN","Název"]).size().sort_values(ascending=False).to_frame().reset_index()

ISBN_nemame.info()
ISBN_nemame.head()
```

# ISSN

```python
ISSN_result = pd.merge(Citace_vyber, ISSN_MU[["norma","zdroj"]], left_on = "ISSN_norma", right_on = "norma", how="left")
ISSN_result.head()
```

```python
#kolik je nalezených shod a co nejčastě citují z toho co máme a nemáme
ISSN_result.groupby(["zdroj"]).size()
```

```python
#co máme ve fondu
ISSN_notNan = ISSN_result[ISSN_result["zdroj"].notnull() & ISSN_result["ISSN"].notnull()]
ISSN_mame = ISSN_notNan.groupby(["ISSN","Název"]).size().sort_values(ascending=False).to_frame().reset_index()

ISSN_mame.info()
ISSN_mame.head(20)
```

```python
#co nemáme ve fondu
ISSN_Nan = ISSN_result[ISSN_result["zdroj"].isna() & ISSN_result["ISSN"].notnull()]
ISSN_nemame = ISSN_Nan.groupby(["ISSN","Název"]).size().sort_values(ascending=False).to_frame().reset_index()

ISSN_nemame.info()
ISSN_nemame.head(20)
```

```python
#vytvoření souboru xlsx s výstupy

dfs = {'ISBN_all': ISBN_result, 'ISBN máme': ISBN_mame, 'ISBN nemáme': ISBN_nemame, 'ISSN_all': ISSN_result, 'ISSN máme' : ISSN_mame, 'ISSN nemáme' : ISSN_nemame}

writer = pd.ExcelWriter("Srovnani_MU_Citace.xlsx", engine='xlsxwriter')
for sheetname, df in dfs.items():  # loop through `dict` of dataframes
    df.to_excel(writer, sheet_name=sheetname, index=False)  # send df to writer
    worksheet = writer.sheets[sheetname]  # pull worksheet object
    for idx, col in enumerate(df):  # loop through all columns
        series = df[col]
        max_len = max((
            series.astype(str).map(len).max(),  # len of largest item
            len(str(series.name))  # len of column name/header
            )) + 1  # adding a little extra space
        worksheet.set_column(idx, idx, max_len)  # set column width
        #sheet = worksheet
        #sheet["A1:"].alignment = left_aligned_text
        #sheet["A1:"].alignment = text_rotation
        
writer.save()
```
